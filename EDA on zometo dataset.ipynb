```python
def handlerate(value):
    if(value=='NEW'or value=="-"):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)   
    
```

```python
df['rate']=df['rate'].apply(handlerate)
```

```python
df['rate'].head()
```

```python
df['rate'].fillna(df['rate'].mean(),inplace=True)
```

```python
df['rest_type'].value_counts()
```

```python
rest_types=df['rest_type'].value_counts()
```

```python
rest_type_lessthan1000 = rest_types[rest_types<1000]
rest_type_lessthan1000
```

```python
def handel_rest_type(value):
    if(value in rest_type_lessthan1000):
        return "others"
    else:
        return value
```

```python
df['rest_type']= df['rest_type'].apply(handel_rest_type)
df['rest_type'].value_counts()
```

```python
location = df['location'].value_counts()
```

```python
location_lessthan300 = location[location<300]
```

```python
def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
```

```python
df['location']=df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
df.rename(columns={'approx_cost(for two people)':'Cost2plates','listed_in(type)':'Type'},inplace=True)
```

```python
df.head()
```

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)
    
df['Cost2plates'] = df['Cost2plates'].apply(handlecomma)
df['Cost2plates'].unique()
```
